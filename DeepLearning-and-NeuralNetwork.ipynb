LogisticRegression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("Bank.CSV")
data.head()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
x=data.iloc[:,:7]
y=data.iloc[:,7]
```

```python
x_train,x_tes,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
clf = MLPClassifier(alpha=0.01,hidden_layer_sizes=(5,3),activation="logistic")
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_tes)
```

```python
accuracy_score(y_test,y_pred)
```

tensorflow and keras

Classification

```python
import tensorflow
from keras.models import Sequential
from keras.layers import InputLayer,Dense
from keras.optimizers import Adam
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(7,)))
model.add(Dense(units=5,activation="relu"))
model.add(Dense(units=3,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
```

```python
model.summary()
```

```python
optim = Adam(learning_rate=0.01)
```

```python
model.compile(loss="binary_crossentropy",optimizer=optim)
```

```python
model.fit(x_train,y_train,epochs=50)
```

```python
y_pred = np.argmax(model.predict(x_tes), axis=1)
```

```python
accuracy_score(y_test,y_pred)
```

Convolutional Neural Network

```python
from tensorflow.keras.datasets import mnist
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mnist.load_data()[0][0][1])
plt.show()
```

```python
plt.imshow(mnist.load_data()[0][0][1],cmap="gray")
plt.show()
```

```python
mnist.load_data()[0][1][1]
```

```python
(x_train,y_train) , (x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(60000,28,28,1)
x_test=x_test.reshape(10000,28,28,1)
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation="relu",input_shape=(28,28,1)))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(100,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=20)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```
