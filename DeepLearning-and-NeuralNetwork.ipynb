```python
training_set_scaled
```

```python
n_steps = 60

x=[]
y=[]
for i in range(len(training_set_scaled)):
    end_ix = i + n_steps
    if end_ix > len(training_set_scaled) - 1:
        break
    seq_x, seq_y = training_set_scaled[i:end_ix], training_set_scaled[end_ix]
    x.append(seq_x)
    y.append(seq_y)

x_train=np.array(x)
y_train=np.array(y)
```

LSTM

```python
from tensorflow.keras.layers import Dense, LSTM
```

```python
model_lstm = Sequential()
model_lstm.add(LSTM(units=125, activation="tanh", input_shape=(n_steps, 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer="RMSprop", loss="mse")
model_lstm.summary()
```

```python
model_lstm.fit(x_train, y_train, epochs=50)
```

```python
dataset_total = data.loc[:,"High"]

inputs = dataset_total[len(dataset_total) - len(test) - n_steps :].values
inputs = inputs.reshape(-1, 1)
inputs = sc.transform(inputs)

n_steps = 60

x=[]
y=[]
for i in range(len(inputs)):
    end_ix = i + n_steps
    if end_ix > len(inputs) - 1:
        break
    seq_x, seq_y = inputs[i:end_ix], inputs[end_ix]
    x.append(seq_x)
    y.append(seq_y)

x_test=np.array(x)
y_test=np.array(y)
```

```python
import tensorflow
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train,x_test = x_train/255.0, x_test/255.0
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
plt.matshow(x_test[56])
print(y_test[56])
```

```python
x_train_flatten = x_train.reshape(60000,-1)
x_test_flatten = x_test.reshape(10000, -1)
```

```python
from tensorflow import keras
```

```python
y_train_categorical = keras.utils.to_categorical(y_train,num_classes=10)
y_test_categorical=keras.utils.to_categorical(y_test,num_classes=10)
print(y_train_categorical.shape)
print(y_test_categorical.shape)
```

```python
w = np.zeros((784,10))
b = np.zeros((1,10))
```

```python
learning_rate = 0.001
mse_list=[]
epochs_list=[]
accuracy_list=[]
epochs = 1000
```

```python
for i in range(epochs):
    y_pred =np.dot(x_train_flatten,w) + b
    error = y_train_categorical - y_pred

    grad_w = -2 * (np.dot(x_train_flatten.T,error)/len(x_train_flatten))
    grad_b = -2 * np.mean(error,axis=0,keepdims=True)

    w = w - learning_rate * grad_w
    b = b - learning_rate * grad_b

    predicted_labels = np.argmax(y_pred, axis=1)
    actual_labels = np.argmax(y_train_categorical, axis=1)

    accuracy = np.mean(predicted_labels == actual_labels) * 100
    accuracy_list.append(accuracy)

    mse_list.append(np.mean(np.square(error)))
    epochs_list.append(i)
    print(accuracy)
```

```python
plt.plot(epochs_list,mse_list,label="MSE")
```

```python
plt.plot(epochs_list,accuracy_list,label="Accuracy")
```

```python
y_pred=np.dot(x_test_flatten,w)+b
pred_labels = np.argmax(y_pred, axis=1)
actual_labels = np.argmax(y_test_categorical, axis=1)
accuracy=np.mean(pred_labels == actual_labels) * 100
print(accuracy)
```

ANN

```python
from keras.datasets import mnist
import torch
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train/255.0
x_test = x_test/255.0

x_train = x_train.reshape(60000,28*28)
x_test = x_test.reshape(10000,28*28)
print(x_train.shape)
print(y_test.shape)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
x_train = torch.tensor(x_train,dtype=torch.float32).to(device)
x_test = torch.tensor(x_test,dtype=torch.float32).to(device)
y_train = torch.tensor(y_train,dtype=torch.long).to(device)
y_test = torch.tensor(y_test,dtype=torch.long).to(device)
```

```python
dataset = TensorDataset(x_train,y_train)
dataloader = DataLoader(dataset,batch_size=54,shuffle=True)
for batch in dataloader:
    print(batch)
```

```python
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(28*28, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 10)
        self.softmax = nn.LogSoftmax(dim=1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        return x
```

```python
import torch.optim as optim
```

```python
model = ANN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 15
for epoch in range(epochs):

    running_loss = 0.0
    correct = 0
    total = 0

    for data, target in dataloader:
        pred = model(data)
        loss = criterion(pred, target)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        running_loss += loss.item()

        _, predicted = torch.max(pred, 1)
        correct += (predicted == target).sum().item()
        total += target.size(0)

    accuracy = 100 * correct / total
```

```python
test_dataset = TensorDataset(x_test,y_test)
test_dataloader = DataLoader(test_dataset,batch_size=54,shuffle=True)
```

```python
correct_test = 0
total_test = 0
with torch.no_grad():
    for images, labels in test_dataloader:
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        correct_test += (predicted == labels).sum().item()
        total_test += labels.size(0)

    test_accuracy = 100 * correct_test/total_test

    print(f"Loss: {running_loss / len(test_dataloader):.4f}, test_accuracy:{test_accuracy:.2f}")
```
