LogisticRegression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("Bank.CSV")
data.head()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
x=data.iloc[:,:7]
y=data.iloc[:,7]
```

```python
x_train,x_tes,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
clf = MLPClassifier(alpha=0.01,hidden_layer_sizes=(5,3),activation="logistic")
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_tes)
```

```python
accuracy_score(y_test,y_pred)
```

LinearRegression

```python
data=pd.read_csv("Housing Price.csv")
data.head()
```

```python
x=data.iloc[:,:5].values
y=data.iloc[:,5].values
```

```python
x_train,x_tes,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
rg = MLPRegressor(alpha=0.01,hidden_layer_sizes=(3,2),activation="identity")
```

```python
rg.fit(x_train,y_train)
```

```python
y_pred=rg.predict(x_tes)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
data=pd.read_csv("Bank.CSV")
data.head()
```

```python
x=data.iloc[:,:7]
y=data.iloc[:,7]
```

```python
x_train,x_tes,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

tensorflow and keras

Classification

```python
import tensorflow
from keras.models import Sequential
from keras.layers import InputLayer,Dense
from keras.optimizers import Adam
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(7,)))
model.add(Dense(units=5,activation="relu"))
model.add(Dense(units=3,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
```

```python
model.summary()
```

```python
optim = Adam(learning_rate=0.01)
```

```python
model.compile(loss="binary_crossentropy",optimizer=optim)
```

```python
model.fit(x_train,y_train,epochs=50)
```

```python
y_pred = np.argmax(model.predict(x_tes), axis=1)
```

```python
accuracy_score(y_test,y_pred)
```
