```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("Bank.CSV")
data.head()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
x=data.iloc[:,:7]
y=data.iloc[:,7]
```

```python
x_train,x_tes,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
clf = MLPClassifier(alpha=0.01,hidden_layer_sizes=(5,3),activation="logistic")
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_tes)
```

```python
accuracy_score(y_test,y_pred)
```

Convolutional Neural Network

```python
from tensorflow.keras.datasets import mnist
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mnist.load_data()[0][0][1])
plt.show()
```

```python
plt.imshow(mnist.load_data()[0][0][1],cmap="gray")
plt.show()
```

```python
mnist.load_data()[0][1][1]
```

```python
(x_train,y_train) , (x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(60000,28,28,1)
x_test=x_test.reshape(10000,28,28,1)
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation="relu",input_shape=(28,28,1)))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(100,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=20)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```

Recurrent Neural Network

```python
data=pd.read_csv("Mastercard_stock.CSV",index_col="Date")
data.head()
```

```python
data.loc[:,"High"].plot(figsize=(16,4),legend=True)
plt.show()
```

```python
tstart=2016
tend=2020
data.loc["2016":"2020","High"].plot(figsize=(16,4),legend=True)
plt.show()
```

```python
data.loc["2021":,"High"].plot(figsize=(16,4),legend=True)
plt.show()
```

```python
train=data.loc["2016":"2020","High"].values
test=data.loc["2021":,"High"].values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler(feature_range=(0,1))
training_set=train.reshape(-1,1)
training_set_scaled=sc.fit_transform(training_set)
```

```python
training_set_scaled
```

```python
n_steps = 60

x=[]
y=[]
for i in range(len(training_set_scaled)):
    end_ix = i + n_steps
    if end_ix > len(training_set_scaled) - 1:
        break
    seq_x, seq_y = training_set_scaled[i:end_ix], training_set_scaled[end_ix]
    x.append(seq_x)
    y.append(seq_y)

x_train=np.array(x)
y_train=np.array(y)
```

```python
from tensorflow.keras.layers import Dense, LSTM
```

```python
model_lstm = Sequential()
model_lstm.add(LSTM(units=125, activation="tanh", input_shape=(n_steps, 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer="RMSprop", loss="mse")
model_lstm.summary()
```

```python
model_lstm.fit(x_train, y_train, epochs=50)
```

```python
dataset_total = data.loc[:,"High"]

inputs = dataset_total[len(dataset_total) - len(test) - n_steps :].values
inputs = inputs.reshape(-1, 1)
inputs = sc.transform(inputs)

n_steps = 60

x=[]
y=[]
for i in range(len(inputs)):
    end_ix = i + n_steps
    if end_ix > len(inputs) - 1:
        break
    seq_x, seq_y = inputs[i:end_ix], inputs[end_ix]
    x.append(seq_x)
    y.append(seq_y)

x_test=np.array(x)
y_test=np.array(y)
```

```python
predicted_stock_price = model_lstm.predict(x_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
plt.plot(data.loc["2021":,"High"].values,color="gray",label="Real")
plt.plot(predicted_stock_price,color="red",label="Predicted")
plt.title("MasterCard Stock Price Prediction")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```

```python
np.sqrt(mean_squared_error(data.loc["2021":,"High"].values,predicted_stock_price))
```
