Recurrent Neural Network

```python
data=pd.read_csv("Mastercard_stock.CSV",index_col="Date")
data.head()
```

```python
data.loc[:,"High"].plot(figsize=(16,4),legend=True)
plt.show()
```

```python
tstart=2016
tend=2020
data.loc["2016":"2020","High"].plot(figsize=(16,4),legend=True)
plt.show()
```

```python
data.loc["2021":,"High"].plot(figsize=(16,4),legend=True)
plt.show()
```

```python
train=data.loc["2016":"2020","High"].values
test=data.loc["2021":,"High"].values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler(feature_range=(0,1))
training_set=train.reshape(-1,1)
training_set_scaled=sc.fit_transform(training_set)
```

```python
training_set_scaled
```

```python
n_steps = 60

x=[]
y=[]
for i in range(len(training_set_scaled)):
    end_ix = i + n_steps
    if end_ix > len(training_set_scaled) - 1:
        break
    seq_x, seq_y = training_set_scaled[i:end_ix], training_set_scaled[end_ix]
    x.append(seq_x)
    y.append(seq_y)

x_train=np.array(x)
y_train=np.array(y)
```

```python
from tensorflow.keras.layers import Dense, LSTM
```

```python
model_lstm = Sequential()
model_lstm.add(LSTM(units=125, activation="tanh", input_shape=(n_steps, 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer="RMSprop", loss="mse")
model_lstm.summary()
```

```python
model_lstm.fit(x_train, y_train, epochs=50)
```

```python
dataset_total = data.loc[:,"High"]

inputs = dataset_total[len(dataset_total) - len(test) - n_steps :].values
inputs = inputs.reshape(-1, 1)
inputs = sc.transform(inputs)

n_steps = 60

x=[]
y=[]
for i in range(len(inputs)):
    end_ix = i + n_steps
    if end_ix > len(inputs) - 1:
        break
    seq_x, seq_y = inputs[i:end_ix], inputs[end_ix]
    x.append(seq_x)
    y.append(seq_y)

x_test=np.array(x)
y_test=np.array(y)
```

```python
predicted_stock_price = model_lstm.predict(x_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
plt.plot(data.loc["2021":,"High"].values,color="gray",label="Real")
plt.plot(predicted_stock_price,color="red",label="Predicted")
plt.title("MasterCard Stock Price Prediction")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```

```python
np.sqrt(mean_squared_error(data.loc["2021":,"High"].values,predicted_stock_price))
```

Artifical Neural Network

```python
import tensorflow
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train,x_test = x_train/255.0, x_test/255.0
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
plt.matshow(x_test[56])
print(y_test[56])
```

```python
x_train_flatten = x_train.reshape(60000,-1)
x_test_flatten = x_test.reshape(10000, -1)
```

```python
from tensorflow import keras
```

```python
y_train_categorical = keras.utils.to_categorical(y_train,num_classes=10)
y_test_categorical=keras.utils.to_categorical(y_test,num_classes=10)
print(y_train_categorical.shape)
print(y_test_categorical.shape)
```

```python
w = np.zeros((784,10))
b = np.zeros((1,10))
```

```python
learning_rate = 0.001
mse_list=[]
epochs_list=[]
accuracy_list=[]
epochs = 1000
```

```python
for i in range(epochs):
    y_pred =np.dot(x_train_flatten,w) + b
    error = y_train_categorical - y_pred

    grad_w = -2 * (np.dot(x_train_flatten.T,error)/len(x_train_flatten))
    grad_b = -2 * np.mean(error,axis=0,keepdims=True)

    w = w - learning_rate * grad_w
    b = b - learning_rate * grad_b

    predicted_labels = np.argmax(y_pred, axis=1)
    actual_labels = np.argmax(y_train_categorical, axis=1)

    accuracy = np.mean(predicted_labels == actual_labels) * 100
    accuracy_list.append(accuracy)

    mse_list.append(np.mean(np.square(error)))
    epochs_list.append(i)
    print(accuracy)
```

```python
plt.plot(epochs_list,mse_list,label="MSE")
```

```python
plt.plot(epochs_list,accuracy_list,label="Accuracy")
```

```python
y_pred=np.dot(x_test_flatten,w)+b
pred_labels = np.argmax(y_pred, axis=1)
actual_labels = np.argmax(y_test_categorical, axis=1)
accuracy=np.mean(pred_labels == actual_labels) * 100
print(accuracy)
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train,x_test = x_train/255.0, x_test/255.0
```

```python
x_train_flatten = x_train.reshape(60000,-1)
x_test_flatten = x_test.reshape(10000, -1)
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(units=128, activation="relu"),
    keras.layers.Dense(units=10, activation="relu"),
    keras.layers.Dense(units=10,activation="softmax")
])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train_flatten,y_train,epochs=10,validation_split=0.2)
```

```python
print(history.history.keys())
```

```python
model.evaluate(x_test_flatten,y_test)
```

```python
plt.plot(history.history['accuracy'],label="accuracy")
plt.plot(history.history['val_accuracy'],label="val_accuracy")
plt.legend()
```

```python
plt.plot(history.history['loss'],label="Loss")
plt.plot(history.history['val_loss'],label="Validation Loss")
plt.legend()
```
